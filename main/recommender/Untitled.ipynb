```python
import pandas as pd
```

```python
df = pd.read_csv('models/movie_based_collaborative_filtering.csv', index_col=0)
```

```python
df
```

```python
CORR_MATRIX = df
def get_similar_movies(movie_name, rating):
    similar_ratings = CORR_MATRIX[movie_name] * (rating - 2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    return 


def predict(list_of_movies):
    predictions = []
    for movie, rating in list_of_movies:
        predictions.append(get_similar_movies(movie, rating))
    return predictions
```

```python
x = [("12 Angry Men (1957)", 4)]

print(predict(x))
```
